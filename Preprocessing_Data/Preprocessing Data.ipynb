```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 

dataset = pd.read_csv('Dataset_Banjir.csv') 
x = dataset.iloc[:, :-1].values # untuk atribut
y = dataset.iloc[:, -1].values  # untuk class
```

```python
print(x)
```

```python
print(y)
```

```python
# untuk mengisi missing value (nan) dengan nilai rata2 (mean)
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:, 1:4])
x[:, 1:4] = imputer.transform(x[:, 1:4])
```

```python
print(x)
```

```python
#function untuk mengubah baris awal string menjadi angka
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
print(x)
```

```python
from sklearn.preprocessing import LabelEncoder # function utk mengubah label yg yes no menjadi numerik
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=1) 
# 0.2 artinya 20% utk test, dan sisanya 80% untuk train
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[:, 8:] = sc.fit_transform(x_train[:, 8:])
x_test[:, 8:] = sc.transform(x_test[:, 8:])
```
